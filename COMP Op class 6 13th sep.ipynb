```python
# what if i want to create 1000 uniformly random directions?
import scipy.linalg
norms = scipy.sqrt((xy2*xy2).sum(axis=1))
norms.shape
```

```python
xy2 = xy2/norms.reshape(1000,1)
```

```python
pylab.scatter(xy2[:,0],xy2[:,1])
```
